# Добро пожаловать на курс по этике ИИ!

ИИ все сильнее влияет на все - от социальных сетей до здравоохранения. ИИ используется для принятия решений по кредитным картам, для видеонаблюдения в аэропортах и для информирования военных операций. Эти технологии могут как вредить, так и помогать людям, которым они служат. Применяя этическую оптику, мы можем стремиться выявлять вред, который эти технологии могут причинять людям, и проектировать и создавать их так, чтобы уменьшать этот вред - или решать не создавать их вовсе.

**Для этого курса не требуется предварительной подготовки**, и мы не предполагаем наличия опыта программирования.

Ваши преподаватели **[Var Shankar](https://www.kaggle.com/var0101)** (который работал над этикой ИИ в индустрии) и **[Alexis Cook](https://www.kaggle.com/alexisbcook)** (которая проектировала и обучала модели ИИ) совместно создали этот прикладной и практический курс. В упражнениях вы будете запускать код для обучения и исследования моделей ИИ. Если вы уже работаете в команде, которая разрабатывает ИИ-продукты, и не знаете, с чего начать, мы даем практические рекомендации.

Этот курс охватывает несколько тем:
- На уроке **о дизайне, ориентированном на человека**, вы узнаете, как проектировать систему ИИ так, чтобы она служила потребностям людей, для которых она предназначена.
- На уроке **о смещениях (bias)** вы определите, как системы ИИ могут научиться дискриминировать определенные группы.
- На уроке **о справедливости (fairness)** вы научитесь количественно оценивать степень смещения в системах ИИ.
- На уроке **о model cards** вы узнаете, как использовать популярный фреймворк для повышения общественной подотчетности моделей ИИ.

По мере развития области ИИ развивается и этика ИИ. Как однажды сказал наш мудрый друг, _"Этика - это разговор."_ Этот курс может помочь вам начать этот разговор, но не завершить его. Некоторые темы, которые мы не рассматриваем, включают потенциальное влияние ИИ на цифровой разрыв между богатыми и бедными, на рабочие места и на возможности авторитарных правительств, среди прочего. Тем не менее на протяжении курса мы рекомендуем следующие шаги для продолжения обучения и отслеживания этического разговора по мере его развития.
